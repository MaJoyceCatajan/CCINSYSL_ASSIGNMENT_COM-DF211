2. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | CT | DG | PPL | BNN | RNG


```python
# Answer here

def vow():
  input = 'ORANGE'
  vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I' , 'O', 'U']
  blank = ''

  for char in input:
    if char not in vowels:
      blank = blank + char
  print (blank)



vow()
```
